```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Считаем все виды расстояний между парами точек
def print_distances(p_arr):
    for i in range(len(p_arr)):
        for j in range(i + 1, len(p_arr)):
            print('p' + str(i + 1), 'p' + str(j + 1))

            print("euclidean",          np.linalg.norm(p_arr[i] - p_arr[j]))
            print("square euclidean",   np.linalg.norm(p_arr[i] - p_arr[j])**2)
            print("chebyshev",          np.linalg.norm(p_arr[i] - p_arr[j], ord = np.inf))
            print("hemming",            np.linalg.norm(p_arr[i] - p_arr[j], ord = 1), '\n')


# Точки
p1 = np.array([1, 2, 3])
p2 = np.array([3, 6, 9])
p3 = np.array([5, 10, 0])
p4 = np.array([10, 10, 0])

# Создаём график
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

# Добавляем точки на график
ax.scatter(p1[0], p1[1], p2[2])
ax.scatter(p2[0], p2[1], p2[2])
ax.scatter(p3[0], p3[1], p3[2])
ax.scatter(p4[0], p4[1], p4[2])

p_arr = np.array([p1, p2, p3, p4])

print_distances(p_arr)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns

iris = sns.load_dataset('iris')
iris
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16,7))

plt.subplot(121)
sns.scatterplot(
    data=iris,
    x='petal_width', y='petal_length',
    hue='species',
    s=70
)
plt.xlabel('Длина лепестка, см')
plt.ylabel('Ширина лепестка, см')
plt.legend()
plt.grid()

plt.subplot(122)
sns.scatterplot(
    data=iris,
    x='sepal_width', y='sepal_length',
    hue='species',
    s=70
)
plt.ylabel('Длина чашелистика, см')
plt.xlabel('Ширина чашелистика, см')
plt.legend()
plt.grid()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:, :-1], iris.iloc[:, -1], test_size=0.20)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
import numpy as np

plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=iris,
    x='petal_width', y='petal_length',
    hue='species',
    s=70
)
plt.ylabel('Длина лепестка, см')
plt.xlabel('Ширина лепестка, см')
plt.legend()
plt.grid()

for i in range(len(y_test)):
    if np.array(y_test)[i] != y_pred[i]:
        plt.scatter(X_test.iloc[i, 3], X_test.iloc[i, 2], color='red', s=150)
```

```python
from sklearn.metrics import accuracy_score
print(f'accuracy: {accuracy_score(y_test, y_pred) :.3}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:, :-1], iris.iloc[:, -1], test_size = 0.15)

model1 = KNeighborsClassifier(n_neighbors = 1)
model2 = KNeighborsClassifier(n_neighbors = 5)
model3 = KNeighborsClassifier(n_neighbors = 10)
```

```python
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model3.fit(X_train, y_train)

y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)
y_pred3 = model3.predict(X_test)
```
