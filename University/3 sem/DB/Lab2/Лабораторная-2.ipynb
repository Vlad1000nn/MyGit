## Задание 1
Вывести названия всех групп, их песен и названия их альбомов для всех треков жанра Rock, приобретенные сотрудниками Microsoft.


```python
%%sql

SELECT ar.Name as GroupName, t.Name as TrackName, al.Title as AlbumName FROM Genres g

JOIN Tracks          t  ON t.GenreId    = g.GenreId
JOIN Invoice_items   ii ON ii.TrackId   = t.TrackId 
JOIN Albums          al ON al.AlbumId   = t.AlbumId
JOIN Artists         ar ON ar.ArtistId  = al.ArtistId
JOIN Invoices        i  ON i.InvoiceId  = ii.InvoiceId
JOIN Customers       c  ON i.CustomerId = c.CustomerId

WHERE g.Name = 'Rock' AND c.Company = 'Microsoft Corporation'
```

## Задание 2
Вывести 10 самых больших по размеру треков жанра Blues и формата MPEG (используйте команду LIMIT 10). Только название и автор.


```python
%%sql
SELECT t.Name as TrackName,ar.Name as Author FROM Tracks t

JOIN Genres      g  ON g.GenreId     = t.GenreId
JOIN Media_Types mt ON t.MediaTypeId = mt.MediaTypeId
JOIN Albums      al ON al.AlbumId    = t.AlbumId
JOIN Artists     ar ON ar.ArtistId   = al.ArtistId

WHERE mt.Name = 'MPEG audio file' AND g.Name = 'Blues'

ORDER BY t.Bytes DESC

LIMIT 10
```

## Задание 3
Для каждой компании вывести общее количество купленных джазовых композиций.


```python
%%sql

SELECT c.Company, count(c.Company) as count FROM Customers c
--SELECT * FROM Customers c

JOIN Invoices       i   ON i.CustomerId = c.CustomerId
JOIN Invoice_Items  ii  ON ii.InvoiceId = i.InvoiceId
JOIN Tracks         t   ON t.TrackId    = ii.TrackId
JOIN Genres         g   ON g.GenreId    = t.GenreId

WHERE g.Name = 'Jazz' AND c.Company IS NOT null

GROUP BY c.Company
```

## Задание 4
Вывести топ 3 альбома по продолжительности (продолжительность альбома - сумма длин треков в нём). Только название и продолжительность.


```python
%%sql

--SELECT * FROM Albums al
SELECT al.Title as AlbumName, sum(t.Milliseconds) as duration FROM Albums al

JOIN Tracks t ON al.AlbumId = t.AlbumId

GROUP BY al.Title

ORDER BY duration DESC

LIMIT 3
```

## Задание 5
Для каждого набора (жанр, тип медиа) вывести среднюю цену по стоимости трека и их общее количество, причем вывести только те наборы, для которых **все** треки стоят больше 1,5$ и для которых в наборе есть хотя бы один трек.

```python
%%sql

SELECT g.Name as Genre, mt.Name as MediaType, avg(t.UnitPrice) as average, count(*) as count FROM Genres g

JOIN Tracks      t  ON t.GenreId      = g.GenreId
JOIN Media_Types mt ON mt.MediaTypeId = t.MediaTypeId

GROUP BY g.Name, mt.Name
having min(t.UnitPrice) > 1.5 AND count > 0
```
